```python
# Web Scraping (e.g., BeautifulSoup)
import requests
from bs4 import BeautifulSoup
# Specify the URL of the website
url = "https://books.toscrape.com/"
# Send a GET request
response = requests.get(url)
# Check for successful response
if response.status_code == 200:
# Parse the HTML content
 	soup = BeautifulSoup(response.content, 'html.parser')
# Find all product elements (articles)
products = soup.find_all('article')
# Extract and print book titles
for product in products:
	title_element = product.h3.a
	if title_element: # Check if title element exists
		book_title = title_element.text.strip()
		print(book_title)
	else:
		print(f"Error: Failed to download page. Status code: {response.status_code}")
# products
```
